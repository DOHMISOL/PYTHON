```python
#컬럼의 타입을 변경시키는 코드다(전체칼럼을 float으로 바꿔보자!)
indata["CUSTOMERCODE"] = indata["CUSTOMERCODE"].astype(float)
```

```python
columnList = list( indata.columns )
columnList
```

```python
for i in range(0,len(columnList)):
    targetColumn = columnList [i]
    print(targetColumn)

    #타입변경(기존타입 -> float)
    try:
        indata [ targetColumn ] = indata [ targetColumn ].astype(float)
    #타입을 못바꾸는 경우(오류메시지 출력), *에러는 발생되지 않음
    except Exception as e : 
        print (e)
```
