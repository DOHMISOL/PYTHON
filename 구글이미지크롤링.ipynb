```python
from selenium import webdriver 
from selenium.webdriver.common.keys import Keys 
from selenium.webdriver.common.by import By
import time 
import os 
import urllib.request
```

```python
def createFolder(directory): 
    try: 
        if not os.path.exists(directory): 
            os.makedirs(directory) 
    except OSError: 
        print ('Error: Creating directory. ' + directory)
```

```python
print(keyword + ' 스크롤 중 .............')
elem = driver.find_element(By.TAG_NAME, "body")
resultView = '//*[@id="islmp"]/div/div/div/div[1]/div[2]/div[2]/input'
for i in range(60):
    elem.send_keys(Keys.PAGE_DOWN)
    time.sleep(0.1)
    try:
        driver.find_element(By.XPATH, resultView).click()
        for i in range(60):
            elem.send_keys(Keys.PAGE_DOWN)
            time.sleep(0.1)
    except:
        pass
```

```python
#islrg > div.islrc > div.isv-r.PNCib.MSM1fd.BUooTd.fT6ABc > a.wXeWr.islib.nfEiy > div.bRMDJf.islir > img
```

```python
links = []
images = driver.find_elements(By.CSS_SELECTOR, "img.rg_i.Q4LuWd")
for image in images:
    if image.get_attribute('src') is not None:
        links.append(image.get_attribute('src'))

print(keyword + ' 찾은 이미지 개수:', len(links))
time.sleep(2)
```

```python
for k,i in enumerate(links): 
    url = i 
    start = time.time() 
    # url내용을 바로 저장 특정경로에 저장
    urllib.request.urlretrieve(url, targetFolder+"/"+keyword+str(k)+".jpg")  
    print(str(k+1)+'/'+str(len(links))+' '+keyword+' 다운로드 중....... Download time : '+str(time.time() - start)[:5]+' 초') 
    print(keyword+' ---다운로드 완료---') 
    
driver.close()
```
