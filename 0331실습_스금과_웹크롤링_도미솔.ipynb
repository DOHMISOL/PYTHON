```python
import requests
import bs4
import pandas as pd
```

```python
htmlObj = bs4.BeautifulSoup(webPageSrc, "html.parser")
```

```python
htmlObj
```

```python
ty_thmb_view = htmlObj.find(name = "div", attrs = {"id" : "ty_thmb_view"} )
ty_thmb_view
```

```python
ul = ty_thmb_view.find(name ="ul")
ul
```

```python
liList = ul.findAll(name ="li")
len(liList)
```

```python
liList[0]
```

```python
name = liList[0].findAll(name="em", attrs={"class": "tx_ko"})
name0 = name[0].text
name1 = name[1].text

for i in range(len(name)):
    names = name[i].text
    print(names)

price = liList[0].findAll(name="em", attrs={"class": "ssg_price"})
price0 = price[0].text
price1 = price[1].text

for i in range(len(price)):
    pricess = price[i].text
    print(pricess)
```

```python
dfList = []
for i in range(len(liList)):
    name=liList[i].findAll(name ="em", attrs = {"class" : "tx_ko"})
    price = liList[i].findAll(name = "em", attrs = {"class" : "ssg_price"})
    
    rowList= []
    for i in range(len(name)):
        rowList.append(name[i].text)

    rowList1 = []
    for i in range(len(price)):
        rowList1.append(price[i].text)
        
    sumList = rowList + rowList1
    dfList.append(sumList)
# 계속 돌면서 넣는값이 누적이 되지 않기때문에 리스트를 생성하여 만들어준다.


dfList
```

```python
dfList = pd.DataFrame(dfList)
dfList.columns = ["브랜드","이름","할인후가격","할인전가격"]
dfList
```
